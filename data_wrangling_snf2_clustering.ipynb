```python
# load data
import pandas as pd
cluster1 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/snf2/subfamily_cluster_1.csv')
cluster1
```

Rename and drop the column of each cluster:

```python
# rename 'Unnamed: 0','frequency', 'percentage'
cluster1 = cluster1.rename(columns={'Unnamed: 0':'cluster', 'frequency':'cluster 1', 'percentage':'%1'})
cluster1.drop('cluster', inplace=True, axis=1)
cluster1
```

```python
# rename 'Unnamed: 0','frequency', 'percentage'
cluster2 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/snf2/subfamily_cluster_2.csv')
cluster2 = cluster2.rename(columns={'Unnamed: 0':'cluster', 'frequency':'cluster 2', 'percentage':'%2'})
cluster2.drop('cluster', inplace=True, axis=1)
cluster2
```

```python
# rename 'Unnamed: 0','frequency', 'percentage'
cluster3 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/snf2/subfamily_cluster_3.csv')
cluster3 = cluster3.rename(columns={'Unnamed: 0':'cluster', 'frequency':'cluster 3', 'percentage':'%3'})
cluster3.drop('cluster', inplace=True, axis=1)
cluster3
```

```python
# rename 'Unnamed: 0','frequency', 'percentage'
cluster4 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/snf2/subfamily_cluster_1.csv')
cluster4 = cluster4.rename(columns={'Unnamed: 0':'cluster', 'frequency':'cluster 4', 'percentage':'%4'})
cluster4.drop('cluster', inplace=True, axis=1)
cluster4
```

```python
# load Neighbour-Joining data
NJ = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/snf2/snf2_NJ.csv')
NJ
```

Merge 4 dataframes on a common column in an outer join:

```python
# merge clusters 1 and 2
result = pd.merge(cluster1, cluster2, how="outer", on='subfamilies')
result['tot_frequency_x'].fillna(result['tot_frequency_y'], inplace=True)  # join 2 columns without NaNs
result.drop('tot_frequency_y', inplace=True, axis=1)  # drop a NaN column 

# merge clusters 1,2,and 3
result = pd.merge(result, cluster3, how="outer", on='subfamilies')
result['tot_frequency_x'].fillna(result['tot_frequency'], inplace=True)
result.drop('tot_frequency', inplace=True, axis=1)

# merge clusters 1,2,3,and 4
result = pd.merge(result, cluster4, how="outer", on='subfamilies')
result['tot_frequency_x'].fillna(result['tot_frequency'], inplace=True)
result.drop('tot_frequency', inplace=True, axis=1)

# take a quick look at the dimensions
result.shape
```

```python
# final dataframe
result
```

Rename and reorder columns

```python
result = result.rename(columns={'tot_frequency_x':'#sequences'})
result = result[['subfamilies','cluster 1','%1','cluster 2','%2','cluster 3','%3','cluster 4','%4','#sequences']]
result
```

Missing values in the result because it doesn't exist:

```python
# check data types
result.dtypes
```

Change [the data type](https://docs.scipy.org/doc/numpy-1.12.0/reference/generated/numpy.dtype.kind.html#numpy.dtype.kind) from float to integer and fill missing values NaNs with 0:

```python
# integer in 'cluster ' column and '#sequences'
for i in range(1,5):
    result['cluster {}'.format(i)] = result['cluster {}'.format(i)].fillna(0).astype(int)
result['#sequences'] = result['#sequences'].fillna(0).astype(int)

# fill NAs
for i in range(1,5):
    result['%{}'.format(i)] = result['%{}'.format(i)].fillna(0)
    
# sort in an ascending order by 'subfamilies'
result.sort_values(by=['subfamilies'], inplace=True)
result
```

Select rows based on multiple conditions:

```python
# rows with percentage > 20 and < 60 in 4 K-means clusters
result.loc[(result['%1'] > 20) & (result['%1'] < 60)|
           (result['%2'] > 20) & (result['%2'] < 60)| 
           (result['%3'] > 20) & (result['%3'] < 60)| 
           (result['%4'] > 20) & (result['%4'] < 60)]
```
