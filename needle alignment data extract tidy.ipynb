```python
import pandas as pd
```

```python
#cons -name snf2cons -auto sequences.fa consensus.fa
#needleall sequences.fa consensus.fa -brief -auto -aformat3 simple -stdout | grep '^#[ |=][1|I|S|G]' > needleall.txt
#for needleall may need to include -gapopen 2
seqpath='/Users/andrew/Desktop/snfsearch5/'
indatafile='snf2all-mafft-needleall.txt'
outdatafile='snf2all-mafft-needledata.txt'
```

```python
#Read in data file
needles={}
fi=open(seqpath+indatafile,'r')
while 1:
    d=fi.readline()
    if not d:
        break
    elif d[:5]=='# 1: ':
        sid=d[5:].strip()
        needles[sid]={}
    else:
        tag=d.split()[1][:-1]
        data=d.split()[2]
        if tag=='Identity':
            needles[sid]['identity']=int(data.split('/')[0])
            needles[sid]['count']=int(data.split('/')[1])
        elif tag=='Similarity':
            needles[sid]['similarity']=int(data.split('/')[0])
            needles[sid]['count']=int(data.split('/')[1])
        elif tag=='Gaps':
            needles[sid]['gaps']=int(data.split('/')[0])
            needles[sid]['count']=int(data.split('/')[1])
        elif tag=='Score':
            needles[sid]['score']=float(data)
fi.close()
dfs=pd.DataFrame.from_dict(needles,orient='index')
print('Starting with {} entries'.format(dfs.shape[0]))
dfs
```

```python
dfs.to_csv(seqpath+outdatafile,sep ='\t')
```
